```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('startup_funding.csv')
df['Investors Name']=df['Investors Name'].fillna('undisclosed')
df['Amount in USD']=df['Amount in USD'].str.replace(',','')
df['Amount in USD'].fillna('0',inplace=True)
df['Amount in USD']=df['Amount in USD'].str.replace('undisclosed','0')
df['Amount in USD']=df['Amount in USD'].str.replace('unknown','0')
df['Amount in USD']=df['Amount in USD'].str.replace('Undisclosed','0')
df['Amount in USD']=df['Amount in USD'].str.replace('14342000+','0')
df['Amount in USD']=df['Amount in USD'].str.replace('NaN','0')
df=df[df['Amount in USD'].str.isdigit()]
df['Amount in USD']=df['Amount in USD'].astype('float')
def investor_name(investor):
    st.title(investor)
    #latest investment by date
    st.subheader('latest_investment')
    investor_latest=df[df['Investors Name'].str.contains(investor)].head()[
        ['Date dd/mm/yyyy', 'Startup Name', 'Investors Name', 'Amount in USD']]
    st.dataframe(investor_latest)
    #big 5 investment
    st.subheader('big_5_investment')
    investment_big=df[df['Investors Name'].str.contains('investor')].groupby('Startup Name')['Amount in USD'].sum().sort_values(ascending=False).head()

    st.dataframe(investment_big)

    #vertical invested on
    verical_series = df[df['Investors Name'].str.contains(investor)].groupby('Industry Vertical')['Amount in USD'].sum()

    st.subheader('Sectors invested in')
    fig1, ax1 = plt.subplots()
    ax1.pie(verical_series, labels=verical_series.index, autopct="%0.01f%%")

    st.pyplot(fig1)


st.sidebar.title('Startup Investmont Analysis')
option=st.sidebar.selectbox('select one',['overall analysis','start up','investor'])

if option=='overall analysis':
    st.title('overall analysis')
elif option=='start up':
    st.sidebar.selectbox('select one', sorted(list(df['Startup Name'].unique())))
    st.sidebar.button('click for analysis')
    st.title('start up analysis')
else:
    name=st.sidebar.selectbox('select one', sorted(list(df['Investors Name'].unique())))
    btn2=st.sidebar.button('click for analysis')
    if btn2:
        investor_name(name)


```
